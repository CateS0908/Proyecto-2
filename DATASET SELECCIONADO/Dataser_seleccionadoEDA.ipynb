Manejo de outliers

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
columnas = df.select_dtypes([np.number]).columns
n = 1
plt.figure(figsize=(15,10), dpi=100)
for columna in columnas:
  plt.subplot(4,4,n)
  n = n+1
  plt.ylabel(columna)
  plt.boxplot(df[columna])
  plt.grid(True)
  plt.tight_layout()
```

Se evidencia comportamiento normal de los datos, sin presencia de datos atípicos.

Visualización del comportamiento de los datos

```python
columnas = df.select_dtypes(include=['float64', 'int']).columns
columnas = columnas.drop(['Conversion', 'CustomerID'])

n = 1
plt.figure(figsize=(15,8), dpi=100)
for columna in columnas:
  plt.subplot(4,4,n)
  n = n+1
  sns.histplot(df, x = columna, stat = 'count', kde=True, color= 'lightseagreen', line_kws = {'linewidth': 4})
  plt.grid(True)
  plt.tight_layout()
```

'Age': se evidencia una interacción con las campañas publicitarias de manera uniforme en el rango de edad desde 20 hasta 70 años.

'Income': comportamiento multimodal, con diversos ingresos lo que indica que hay clientes con diferentes capacidades económicas.

'AdSpend': comportamiento multimodal, se aprecia inversiones desde los 500 aproximadamente hasta los 10000 con fuerte crecimiento al pago por publicidad.

'ClickThorughRate': la medición de la proporción de la frecuencia con la que las personas ven un anuncio es grande hay mucho movimiento hacia la página.

'Conversion Rate': refleja el éxito de una campaña, se puede visualizar un comportamiento simétrico con una creciente respuesta por parte de los clientes.

'Websites Visits' medida del engagement e interés con una tendencia multimodal.

'SocialShares': comportamiento bimodal.

Para los datos que tiene este dataset, se puede abordar diversas problemáticas que están vinculadas con el análisis del rendimiento de las campañas de marketing digital.

- Se puede predecir conversiones (clasificación) en donde se evalúa las características de los clientes y de las campañas predicen con más precisión la conversión, con base a los datos identificar si el cliente realizará la compra.
Los modelos idonéos para esto son:


Regresión Logística, para ppredecir la probabilidad de conversión basada en variables como gasto publicitario, tiempo en sitio web y CTR.

Random Forest o XGBoost: Estos pueden capturar relaciones no lineales y realizar una clasificación mucho más precisa.

Clusterización:

Ideal para segmentación de clientes, con base a las característica segmentarlos con los datos contenidos en interacciones, ingresos, comportamiento en el sitio web y si ha compartido en redes.


Modelos:

K-means: ideal para segmentar clientes según varibales como edad, ingresos, visitas al sitio web, páginas por visita y compras anteriores.

DBSCAN: Permite detectar patrones de comportamiento fuera de los clústeres estándar.

Regresión:

Para la problemática con lo referente a factores que puedan influir en l CTR o el gasto publicitario, con el fin de optimizar las campañas para lograr aumentar las conversiones.

Modelos:

Regresión lineal: es una herramienta que permite analizar el impacto de las variables continuas como el gasto publicitario o el CTR en las conversiones.

Clasificación: Para el problema ¿Qué características predicen si un cliente es más probable que permanezca fiel?

Con ayuda del modelo Árboles de Desición que puede clasificar a los clientes según el riesgo de abandono o fidelidad.

Mejora del gasto publicitario o optimización del gasto publicitario con modelos de Regresión y Clasificación, para calcular el gasto publicitario óptimo para maximixar la conversión o el CTR en función de la herrmaienta o plataforma de publicidad.

Regresión lineal múltiple: permite obtener la relación entre gasto publicitario y el retorno en términos de CTR o conversiones.

Modelos de clasificación: clasificar el retonor de una campaña como alta, media o baja efectividad.
